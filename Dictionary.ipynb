Write a program to store product details in a shopping website catalog. Use a dictionary where product ids are keys and product details (name, price, category etc) are values. Write code to add products, search products, and display products in a category.

```python
products={}
while True:
  inp=int(input("""_______________shopping website catalog________________
  1.add product
  2.search product
  3.display product in a category
  4.break -----------"""))
  match inp:
    case 1:
      product_id=input("enter the product id:")
      name=input("enter product name:")
      price=int(input("enter product price (in rupees):"))
      category=input("enter product category:")
      products[product_id]={"name":name,"price":price,"category":category}
    case 2:
      product_id=input("enter the product id which you want:")
      if product_id in products.keys():
        print("product name:{},price:{},category:{}".format(products[product_id]['name'],products[product_id]['price'],products[product_id]['category']))
      else:
        print("the product id which entered is not match with any product in the catlog")
    case 3:
      matching_products = []
      for product_id, details in products.items():
        if details["category"] == category:
            matching_products.append((product_id, details["name"], details["price"]))

      if matching_products:
        print(f"Products in the {category} category:")
        for product_id, name, price in matching_products:
            print(f"Product ID: {product_id}, Name: {name}, Price: ${price}")
      else:
        print(f"No products found in the {category} category")

    case 4:
      break



```

Write a program to play a word guessing game. Use a list to store dictionary words. Pick a random word, display guess letters, check if letter is in word, and track number of guesses. Display message if user guesses the word.

```python
import random
dictionary_words = ["apple", "banana", "cherry", "orange", "grape", "strawberry", "pineapple"]

def pick_random_word(word_list):
    return random.choice(word_list)

def display_word(word, guessed_letters):
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter
        else:
            display += "_"
    return display

def word_guessing_game():
    word_to_guess = pick_random_word(dictionary_words)
    max_guesses = 6
    guessed_letters = []

    print("Welcome to the Word Guessing Game!")

    while max_guesses > 0:
        print("\nWord to guess:", display_word(word_to_guess, guessed_letters))
        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a valid single letter.")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter.")
            continue

        guessed_letters.append(guess)

        if guess in word_to_guess:
            print("Correct guess!")
            if word_to_guess == display_word(word_to_guess, guessed_letters):
                print("Congratulations! You guessed the word:", word_to_guess)
                break
        else:
            max_guesses -= 1
            print(f"Wrong guess! You have {max_guesses} guesses left.")

    if max_guesses == 0:
        print("Out of guesses! The word was:", word_to_guess)

word_guessing_game()
```

Write a program to store usernames of registered users of a website. Use a set to store unique usernames. Write functions to add new users, check if a username exists, and remove users. Display number of registered users.

```python
usernames= set()
while True:
  inp = int(input("""
  ___________________website registration____________________
  1. add new user
  2. checking
  3.removing
  4. display number of registered users
  5. break """))

  match inp:
   case 1:
      user_name = input("Please enter  username : ")
      usernames.add(user_name)
      print("registered successfully!")
   case 2 :
      user_name = input("enter the username : ")
      if user_name in usernames:
        print("username is existing")
      else:
        print("username not exist")
   case 3 :
    user_name = input("Enter your username : ")
    if user_name in usernames:
      usernames.remove(user_name)
      print("{} removed successfully!".format(user_name))
    else:
      print("username not exist")
   case 4:
    print("number of registered users is {} ".format(len(usernames)))
   case 5:
    break
```

Write a program to store email addresses of subscribers to a newsletter. Use a set to store unique email IDs. Write functions to add new subscribers, unsubscribe emails, and send weekly newsletter to all subscribers.
